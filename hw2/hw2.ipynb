```python
class DataManager():
    def __init__(self,X, Y):
        self.X = X
        self.Y = Y
        self.X = self.normalize()
        
    def normalize(self):
        self.mean = np.mean(self.X, axis=0)
        self.sigma = np.std(self.X, axis=0)
        return (self.X-self.mean)/self.sigma

    def split(self, train_fraction=0.8):
        n_train = int(self.X.shape[0]*train_fraction)
        A = np.random.permutation(self.X.shape[0])
        return self.X[A[:n_train],:], self.Y[A[:n_train]], self.X[A[n_train:],:], self.Y[A[n_train:]]
        
```

```python
class SVM():
    def __init__(self, lam=0.001,dims=5):
        self.dims = 5
        self.W = np.zeros(dims)
        self.b = np.zeros(1)
        self.lam = lam
        
    def cost(self, X, y):
        y_pred = np.dot(X,self.W)+self.b
        cost_a = np.mean(np.maximum(0,1-y*y_pred))
        cost_a += self.lam *np.dot(self.W,self.W)
        
        return cost_a
    
    def gradient(self,X, y):
        y_pred = np.dot(X,self.W)+self.b
        
        idx_1 = np.where(y_pred*y >= 1)[0]
        idx_2 = np.where(y_pred*y < 1)[0]
        
        W_grad = self.lam * self.W * len(idx_1)
        W_grad += self.lam * self.W *len(idx_2) 
        W_grad -= np.dot(y[idx_2],X[idx_2])
        W_grad /= y_pred.shape[0]
        
        b_grad = -np.mean(y[idx_2])
        
        return W_grad, b_grad
    
    def predict(self,X):
        pred = np.sign(np.dot(X,self.W)+self.b)
        
        return pred
    
    def accuracy(self,X,y):
        y_pred = self.predict(X)
        
        return np.where(y_pred==y)[0].shape[0]/y.shape[0]
        
        
```
